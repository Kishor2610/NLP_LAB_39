```python
#Assignment NO: 06
#NAME: Kishor More
#Roll NO: 39
#Title: Implement and visualize dependency parsing of textual input using stanford coreNLP and spaCy library.

import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
texts = ["Gus is learning piano", "Mary plays the violin", "John likes to sing"]

for text in texts:
    doc = nlp(text)

    print(f"Dependency parse visualization for text: '{text}'")
    for token in doc:
        print(f'''
        TOKEN: {token.text}
                =====
                {token.tag_ = }
                {token.head.text = }
                {token.dep_ = }''')
```

```python
displacy.serve(doc, style="dep")
```
